```python
def isDiag(matrix):
    if matrix.shape[0]!= matrix.shape[1]:
        return False
    dim = matrix.shape[0]
    for i in range(dim):
        for j in range(dim):
            if matrix[i,j]!= 0 and i!=j :
                return False
    return True
```

```python
def isUpperTriangle(matrix):
    if matrix.shape[0]!= matrix.shape[1]:
        return False
    dim = matrix.shape[0]
    for i in range(dim):
        for j in range(i):
            if matrix[i,j]!=0:
                return False
    return True
```

```python
print(np.triu(matrix))
```

```python
x = np.array([
    [1,3,3],
    [1,1,0],
    [3,0,2]
])
```

```python
print(isUpperTriangle(x))
```

```python
print(np.triu(x))
```

```python
np.array_equal(x,np.triu(x))
```

```python
def isLowerTriangle(matrix):
    if matrix.shape[0]!= matrix.shape[1]:
        return False
    dim = matrix.shape[0]
    for i in range(dim):
        for j in range(i+1,dim):
            if matrix[i,j]!=0:
                return False
    return True
```
